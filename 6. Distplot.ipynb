# Distplot
A Distplot or distribution plot, depicts the variation in the data distribution. Seaborn Distplot represents the overall distribution of continuous data variables. The Distplot depicts the data by a histogram and a line in combination to it.

### Import Libraries

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

### Load the Dataset

```python
dataset = sns.load_dataset("penguins")
dataset = dataset.dropna()  # drop the missing values
dataset.head(3)
```

## Distplot

#### Draw a Simple Distplot

```python
sns.distplot(dataset['flipper_length_mm'])
```

#### Remove the KDE Plot

```python
sns.distplot(dataset['flipper_length_mm'], kde= False)
```

#### Remove the Hist Plot

```python
sns.distplot(dataset['flipper_length_mm'], hist = False)
```

#### plot rugplot with KDE and Hist

```python
sns.distplot(dataset['flipper_length_mm'], rug = True)
```

#### number of bin

```python
sns.distplot(dataset['flipper_length_mm'], bins = 20)
```

#### Styling the Distplot

```python
sns.set(style = 'darkgrid')                                    # apply dark grid theme

plt.figure(figsize = (16, 9))                                  # set the figure size

rug = {'color':'white', 'height' : 0.09, 'alpha': 2}           # customize the rug plot

hist = {'color': 'blue', 'alpha': 0.60}                        # customize the histogram plot

kde = {'color': 'red', 'lw':4.0, 'alpha': 0.90}                # customize the kde plot

sns.distplot(dataset['flipper_length_mm'], rug = True, rug_kws = rug, hist_kws = hist, kde_kws = kde)
```
