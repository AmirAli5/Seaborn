# Violin Plot
A Violin plot is a method of plotting numeric data. It is similar to a box plot, with the addition of a rotated kernel density plot on each side. Violin plots are similar to box plots, except that they also show the probability density of the data at different values, usually smoothed by a kernel density estimator.

### Import Libraries

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

### Load the Dataset

```python
dataset = sns.load_dataset('mpg').dropna()
dataset.head(3)
```

```python
dataset.cylinders.value_counts()
```

```python
# Filter down to only cars with even number of cylinders
dataset = dataset[dataset.cylinders.isin([4, 6, 8])]
```

```python
dataset.cylinders.value_counts()
```

### Violin Plot

#### Violin Plot With One Variable

```python
sns.violinplot(x = dataset['displacement'])
```

#### Violin Plot With Two Variable

```python
sns.violinplot(x = 'cylinders', y = 'displacement', data = dataset)
```

#### pass hue and take only two category of origin

```python
sns.violinplot(x = 'cylinders', y = 'displacement', hue = 'origin', data = dataset[dataset.origin.isin(['japan', 'europe'])])
```

#### Split = True

```python
# split = True only apply when we have hue two category
dataset = dataset[dataset.origin.isin(['japan', 'europe'])]
sns.violinplot(x = 'cylinders', y = 'displacement', hue = 'origin', data = dataset, split = True)
plt.legend(loc = 2)
```

#### Inner instead of boxplot

```python
# inner
dataset = dataset[dataset.origin.isin(['japan', 'europe'])]
sns.violinplot(x = 'cylinders', y = 'displacement', hue = 'origin', data = dataset, split = True, inner = 'quartiles')
plt.legend(loc = 2)
```

#### Style

```python
sns.set(style = 'darkgrid')  

plt.figure(figsize = (16, 9)) 

sns.violinplot(x = 'cylinders', y = 'displacement', hue = 'origin', data = dataset[dataset.origin.isin(['japan', 'europe'])],
              palette = 'husl')
```
