# Pairplot
A pairplot plot a pairwise relationships in a dataset. The pairplot function creates a grid of Axes such that each variable in data will by shared in the y-axis across a single row and in the x-axis across a single column

### Import Library

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

### Import Dataset

```python
dataset = pd.read_csv("C:\\Users\\amira\\Data Science\\0. Some  Basic Work\\06. Seaborn\\Data\\Iris.csv", index_col = "Id")
dataset.head(3)
```

### Pairplot

#### Simply pairplot

```python
sns.pairplot(dataset)
```

#### Set Species as Target Attribute

```python
sns.pairplot(dataset, hue = 'Species')
```

#### Choose manually attribute to plot

```python
sns.pairplot(dataset, vars = ['SepalLengthCm','SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm'], hue = 'Species')
```

#### Change the color

```python
sns.pairplot(dataset, vars = ['SepalLengthCm','SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm'], hue = 'Species', palette = 'Dark2')
plt.savefig('Graph/3.Pairplot.png')
```

#### Check the linearity 

```python
sns.pairplot(dataset, vars = ['SepalLengthCm','SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm'], hue = 'Species', kind = 'reg')
```
