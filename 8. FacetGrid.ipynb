# FacetGrid
The FacetGrid class is useful when you want to visualize the distribution of a variable or the relationship between multiple variables separately within subsets of your dataset. A FacetGrid can be drawn with up to three dimensions: row, col, and hue. The first two have obvious correspondence with the resulting array of axes; think of the hue variable as a third dimension along a depth axis, where different levels are plotted with different colors.

### Import Libraries

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

### Load the Dataset

```python
dataset = penguins = sns.load_dataset('penguins')
dataset.head(3)
```

### FacetGrid
We can make seprate graph for each category using FacetGrid

#### .map()
we use use map to plot which category like histogram, dictplot etc

```python
g = sns.FacetGrid(dataset, col='island')
g.map(sns.distplot, "flipper_length_mm");
```

```python
g = sns.FacetGrid(dataset, row = 'sex', col='island')
g.map(sns.distplot, "flipper_length_mm");
```

#### .map_dataframe()
by using this we can plot as well and also with axes too

```python
g = sns.FacetGrid(dataset, row = 'sex', col='island')
g.map_dataframe(sns.scatterplot, x='bill_depth_mm', y='bill_length_mm')
```

#### hue

```python
g = sns.FacetGrid(dataset, row = 'sex', col='island', hue='species')
g.map_dataframe(sns.scatterplot, x='bill_depth_mm', y='bill_length_mm')
```

#### Customize with Some Style

```python
# apply dark grid theme
sns.set(style = 'darkgrid')                                    

# set the figure size
plt.figure(figsize = (16, 9))

# Plot FacetGrid 
g = sns.FacetGrid(dataset, row = 'sex', col='island', hue='species', palette='prism')
g.map_dataframe(sns.scatterplot, x='bill_depth_mm', y='bill_length_mm')

# labels Titles, X-axis, Y-axis etc
g.set_titles(row_template='{row_name}', col_template='{col_name} Island')
g.set_axis_labels('bill Dephth (mm)', 'bill Length (mm)')
g.add_legend()

# Save the Graph
plt.savefig("Graph/7.FacetGrid.png", dpi=400)
```
